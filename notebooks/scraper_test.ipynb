```python
import pandas as pd
import json
import numpy as np
import csv
```

```python
df  = pd.read_csv('../data/extract/data_restaurants.csv')
```

```python
df.tail(100)  
```

```python
df['prices'] = df['unitMinPrices'].apply(lambda x: x.split(',')[0].replace('[','$'))
df['restaurant_name'] = df['unitMinPrices'].apply(lambda x: x.split(
    ',')[1].replace('---', ' ').replace('--', ' ').replace('-', ' '))
df['valoration'] = df['unitMinPrices'].apply(lambda x: x.split(',')[2].replace(']',''))
```

```python
df
```

```python
df.drop(['unitMinPrices'], axis=1, inplace=True)
```

```python
df.reindex(columns=['restaurant_name', 'dish_name', 'details', 'categories', 'prices', 'valoration'])
```

```python
df.to_csv('../data/transform/ifood_clean_.csv', sep=',', index=False, encoding='utf-8')
```
